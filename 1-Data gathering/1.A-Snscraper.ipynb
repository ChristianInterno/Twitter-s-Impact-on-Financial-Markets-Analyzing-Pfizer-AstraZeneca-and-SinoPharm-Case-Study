```python
# Imports
import snscrape.modules.twitter as sntwitter
import pandas as pd
import os
import datetime
```

```python
#Pfizer
# Setting variables to be used below
# Setto il numero massimo di tweet che decido di esttrarre 
maxTweets = 10000000

# Creating list to append tweet data to
# Creo una lista nella quale inserirÃ² i tweet
tweets_list2 = []

# Using TwitterSearchScraper to scrape data and append tweets to list
# Utilizzo TwitterSearchScraper per estrarre i tweet con i diversi parametri scelti
for i,tweet in enumerate(sntwitter.TwitterSearchScraper('pfizer lang:"en" since:2021-04-01 until:2021-05-25').get_items()):
    if i>maxTweets:
        break
    tweets_list2.append({
        "data":tweet.date,
        "location":tweet.user.location,
        "id":tweet.id, 
        "testo":tweet.content, 
        "username":tweet.user.username,
        "replyCount":tweet.replyCount,
        "like":tweet.likeCount,
        "retweet_count":tweet.retweetCount,
        "quoteCount":tweet.quoteCount,
        "quote":tweet.quoteCount,
        "source":tweet.source,
        "lang":tweet.lang,
        "url":tweet.url,
        "coordinates":tweet.coordinates,
        "place":tweet.place,
        
        "followersCount":tweet.user.followersCount,
        "friendsCount":tweet.user.followersCount,
        
        "statusesCount":tweet.user.statusesCount,
        "favouritesCount":tweet.user.favouritesCount, 
        "listedCount":tweet.user.listedCount,
        "mediaCount":tweet.user.mediaCount,
    })
```

```python
#converto la lista in un dataframe
```

```python
pfizer1 = pd.DataFrame(tweets_list2)
```

```python
#visualizzo il risultato dell'estrazione
```

```python
pfizer1
```

```python
#salvo quindi il dataframe 
```

```python
pfizer1.to_csv('phizer_finale.csv', sep=',', index=False)
```
