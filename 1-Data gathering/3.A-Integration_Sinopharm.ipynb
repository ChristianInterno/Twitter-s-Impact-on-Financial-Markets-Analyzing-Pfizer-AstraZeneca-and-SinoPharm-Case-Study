# Sentiment

```python
analyzer = SentimentIntensityAnalyzer()
df =[]
for sentence in tweet_sino['testo']:
    vs = analyzer.polarity_scores(sentence)
    vs_df = pd.DataFrame.from_dict(vs.values(), orient='columns')
    df.append(vs_df[0][3])
    
tweet_sino['compound'] = df

tweet_sino['compound'].describe()
```

# Sino Tweet _ hourly analyzes 

```python
sino_count=pd.DataFrame()

# Count tweet by hour
sino_count['N of Tweet']= tweet_sino.groupby('Datetime')['id'].count()

# Count like ecc by hour and divide by number of tweet
sino_count['sum of Nlike']= tweet_sino.groupby('Datetime')['like'].sum()
sino_count['median of Nlike']= tweet_sino.groupby('Datetime')['like'].median()
sino_count['avg of Nlike']= tweet_sino.groupby('Datetime')['like'].sum()/sino_count['N of Tweet']
sino_count['max of Nlike']= tweet_sino.groupby('Datetime')['like'].max()#sum()#/sino_count['N of Tweet']

sino_count['sum of retweet']= tweet_sino.groupby('Datetime')['retweet_count'].sum()
sino_count['median of retweet']= tweet_sino.groupby('Datetime')['retweet_count'].median()#sum()#/sino_count['N of Tweet']
sino_count['avg of retweet']= tweet_sino.groupby('Datetime')['retweet_count'].sum()/sino_count['N of Tweet']
sino_count['max of retweet']= tweet_sino.groupby('Datetime')['retweet_count'].max()

sino_count['sum of reply']= tweet_sino.groupby('Datetime')['replyCount'].sum()
sino_count['median of reply']= tweet_sino.groupby('Datetime')['replyCount'].median()#sum()#/sino_count['N of Tweet']
sino_count['avg of reply']= tweet_sino.groupby('Datetime')['replyCount'].sum()/sino_count['N of Tweet']
sino_count['max of reply']= tweet_sino.groupby('Datetime')['replyCount'].max()

sino_count['Comp']= tweet_sino.groupby('Datetime')['compound'].sum()/sino_count['N of Tweet']

sino_count.reset_index(inplace=True)
sino_count
```

```python
# Time delta for time interval
hours= +32   #+24
hours_added = datetime.timedelta(hours = hours)
tweet_sino_c['data']= tweet_sino_c['data'] + hours_added

# Rename
tweet_sino_c = tweet_sino_c.rename(columns={'data': 'Datetime'})

# Sort
tweet_sino_c = tweet_sino_c.sort_values(by="Datetime")

# Delete hour and minutes
tweet_sino_c['date'] = pd.to_datetime(tweet_sino_c['Datetime']).dt.date

# Create column for number of day (0 is Sunday, 6 is Saturday)
tweet_sino_c['day_of_week'] = tweet_sino_c['Datetime'].dt.strftime('%w')
```

```python
# Add a day on for every Sunday
hours=+24
hours_added = datetime.timedelta(hours = hours)

tweet_sino_c['Datetime'] = np.where(
    tweet_sino_c['day_of_week'] == '0',tweet_sino_c['Datetime']+hours_added,tweet_sino_c['Datetime'])

# Add two day on for every Saturday
hours=+48
hours_added = datetime.timedelta(hours = hours)

tweet_sino_c['Datetime'] = np.where(
    tweet_sino_c['day_of_week'] == '6',tweet_sino_c['Datetime']+hours_added,tweet_sino_c['Datetime'])
```
